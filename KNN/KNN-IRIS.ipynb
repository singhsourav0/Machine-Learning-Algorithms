<h1><font color="red">KNN (K Nearest Neighbors) Classification</font></h1>

<font color="blue">K Nearest Neighbors (KNN) is a straightforward machine learning algorithm used for classification tasks. The idea behind KNN is to predict the class of a data point based on the classes of its nearest neighbors.</font>


```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

<font color="green"><b>Sepal length vs Sepal Width (Setosa vs Versicolor)</b></font>

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
```

<font color="green"><b>Petal length vs Petal Width (Setosa vs Versicolor)</b></font>


```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
```

<font color="green"><b>Train Test Split...</b></font>


```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python

len(X_train)
```

```python
len(X_test)
```

<font color="green"><b>Create KNN (K Neighrest Neighbour Classifier)</b></font>


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
knn.predict([[4.8,3.0,1.5,0.3]])
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True, cmap='coolwarm')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
